# Problem 1

Şimdi beraber kullanıcılardan aldığımız verileri işleyerek, onlara hem çıktı verecek, hem de aldığımız ve hesapladığımız verileri ileride kullanabilmek adına kaydedeceğiz.


 **Bu bir veri giriş platformu olacak**


 * kullanıcıya kaç giriş yapacağını soracağız
* kullanıcıdan isim-soyisim, yaş, boy, kilo bilgilerini alarak VKİ hesaplayacağız
* ardından bunları sınıflara göre kaydedeceğiz bir dosyaya kaydedeceğiz


Sınıflar ise şu şekilde: 
* 18, 5 kg/m.'nin altında olanlar: Zayıf
* 18.5 – 24, 9 kg/m. arasında olanlar: Normal kilolu
* 25 – 29, 9 kg/m. arasında olanlar: Fazla kilolu
* 30 – 39, 9 kg/m. arasında olanlar: Obez
* 40 kg/m.'nin üzerinde olanlar: İleri derecede obez (morbid obez), olarak görülür.

```python
import time

def vkiHesaplama(kilo, boy):
    vki= kilo/(boy**2)
    return round(vki,1)

def degerAraligi(vki):
    if vki < 18.5:
        return "zayif"
    elif vki <= 24.9:
        return "normal kilolu"
    elif vki <= 29.9:
        return "fazla kilolu"
    elif vki <= 39.9:
        return "obez"
    elif vki >= 40:
        return "ileri obez"

def ciktiOlustur(isim, yas, boy, kilo, vki, deger):
    kayitTarihi = time.ctime(time.time())
    t1= f"kullanici adi {isim}\n"
    t2= f"kullanicinin yas kilo boy bilgisi {yas}, {kilo}, {boy}\n"
    t3= f"kullanicinin vücut kitle endeksi {vki}\n"
    t4= f"deger araligi {deger}\n"
    t5= f"olusturma tarihi {kayitTarihi}\n"
    output= t1 + t2 + t3 + t4 + t5
    return output

def main():
    girisSayisi= int(input("kaç kullanıcı verisi gireceksiniz: "))
    for i in range(girisSayisi):
        isim = input("isim soyisim giriniz: ")
        yas= input("yasinizi giriniz: ")
        kilo= float(input("kilo giriniz: "))
        boy= float(input("boy giriniz: "))
        vki = vkiHesaplama(kilo, boy)
        deger= degerAraligi(vki)
        output= ciktiOlustur(isim, yas, boy, kilo, vki, deger)

        with open("saglik.txt", "a") as f:
            f.write(output)
            f.write("*"*20)
            f.write("\n")
            print("kullanıcı kaydedildi.")

    print("veriler 'saglik.txt' dosyasina kaydedildi.\n")
main()
```

**YetBar - Smoothie Bank**
---
YetBar spor salonunun girişinde bulunan, insansız bir sağlıklı içecek otomatıdır. Kullanıcı çeşitli girdiler smoothie içeceğine koyacağı ürünleri belirleyebilir, fiyat hesaplaması yaptırabilir ya da sadece menüyü görüntüleyebilir.

**Projenin Amacı:**

*   Birbirini etkileyen ve bir kod içerisinde birden fazla defa başvurulacak fonksiyonların oluşturulup, "fonksiyonel" proglamlamak.
*   Büyük programları yazarken, yazılımcı olarak yapabileceğimiz hatalardan kaçınmak için kontrol yöntemleri geliştirmek.
*   Kullanıcı inputlarını ve yaptığımız hesapları, raporlamak/loglamak amacıyla kaydetmek.

**Programdan Beklenen Çıktılar**
* Program kullanıcıya temelde 2 farklı işlem yapma imkanı sağlar.
> 1. Menüyü görüntüleme
> 3. Satın alma
* Satın alınan her içecek tarih, fiyat bilgisi ile kaydedilmeli ve her zaman ulaşılabilir olmalıdır.




```python
import time

menu = {
    "elma":0.50,
    "armut":0.75,
    "uzum":2.,
    "cilek":1.75,
    "muz":2.,
    "kavun":2.15,
    "karpuz":1.99
}
print("elmanin fiyati: ", menu["elma"])
print("kavunun fiyati: ", menu["kavun"])
def display_menu():
    print("-----YETBAR MENU-----")
    for item in list(menu.keys()):
        print(f"{item.upper():<10} : {menu[item]} TL")

def tek_urun_fiyat_hesaplama(urun, adet):
    assert adet >= 0, "ürün adedi 0 ve 0'dan küçük olamaz."
    if urun not in list(menu.keys()):
        return None
    else:
        return menu[urun] * adet
    
def satin_alma():
    display_menu()
    print("lütfen satın almak istediğiniz ürünü giriniz. çikiş yapmak için 0'ı tuşlayınız.")

    satin_alinan_ürünler = []
    satin_alma_adedi = []
    son_fiyat = 0

    while 1:
        eklenecek_urun = input("hangi üründen eklemek istiyorsunuz? ")
        if eklenecek_urun == "0":
            print("fiş kaydediliyor")
            break
        ekleme_adedi = int(input("kaç adet eklemek istiyorsunuz? "))
        fiyat = tek_urun_fiyat_hesaplama(eklenecek_urun, ekleme_adedi)
        if fiyat is None:
            print("yanlış bir giriş yaptiniz. ürün eklenemedi")
        else:
            print(f"satin alma ücreti: {fiyat}.\nSatin almayı onaylamak için 'y' iptal etmek için 'n' giriniz.")
            check= input("satin almayi onayliyor musunuz? ")
            if check == "y":
                print("satin alim onaylandı.")
                print("-"*40)

                son_fiyat += fiyat
                satin_alinan_ürünler.append(eklenecek_urun)
                satin_alma_adedi.append(ekleme_adedi)
                print(f"ara toplam: {son_fiyat}")
            elif check == "n":
                print("satin alma işlemi iptal edildi")
                print("-"*40)
    return son_fiyat, satin_alinan_ürünler, satin_alma_adedi
def main():
    print("yetbara hoşgeldiniz.!\nMenüyü görüntülemek için 1'e, smoothie yapmak için lütfen 2'ye basiniz. çıkmak için 0 giriniz.")
    program = None
    while 1:
        program = int(input("tercihiniz: "))
        if program < 0 or program > 2:
            print("yanlış bir tercih yaptiniz. lütfen tekrar deneyiniz.")
        elif program == 0:
            print("çıkış yapılıyor")
        elif program == 1:
            display_menu()
        elif program == 2:
            satin_alma_ücreti, satin_alinan_ürünler, satin_alma_adedi = satin_alma()

            with open ("kasa.txt","a") as f:
                f.write(str(time.ctime(time.time())))
                f.write("\n")
                f.write(','.join(map(str, satin_alinan_ürünler)))
                f.write("\n")
                f.write(','.join(map(str, satin_alma_adedi)))
                f.write("\n")
                f.write(f"toplam: {satin_alma_ücreti}")
                f.write("\n")
                f.write("*"*20)
                f.write("\n")

            cikis_kontrol = input("Baska bir islem yapmak istiyor musunuz? (evet-hayir): ")
            
            if cikis_kontrol == "evet":
                pass
            else:
                print("\n\ntesekkürler")
                break
main()
```
