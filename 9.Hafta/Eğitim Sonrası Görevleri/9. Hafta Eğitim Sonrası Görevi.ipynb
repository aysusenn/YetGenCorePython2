# Problem 1

Kumanda isminde bir sınıf oluşturarak aşağıdaki işlemleri yapmasını sağlayan fonksiyonları sınıf içerisinde yazalım.
1. Tv Aç
2. Tv Kapat
3. Ses Ayarları
4. Kanal Ekle
5. Kanal Sayısını Öğrenme
6. Rastgele Kanala Geçme
7. Televizyon Bilgileri


Kumandayı kapatmak için q tuşuna basılmalı ve diğer durumlarda hangi işlemi yapmak istediği sorulmalıdır. Her işlemde gerekli fonksiyonunun sınıf içerisinden çağırılması gerekmektedir.

random ve time kütüphanesini araştırarak kullanabilirsiniz.

Kendi istediğiniz özellikleri eklemekte serbetsiniz.

```python
import random
import time

class Kumanda:
    def __init__(self):
        self.tv_durum = False
        self.ses = 0
        self.kanal_listesi = ["TRT", "FOX", "KANAL D", "TV8", "NTV"]
        self.kanal = "TRT"

    def tv_ac(self):
        if self.tv_durum:
            print("Televizyon zaten açık...")
        else:
            print("Televizyon açılıyor...")
            self.tv_durum = True

    def tv_kapat(self):
        if self.tv_durum:
            print("Televizyon kapanıyor...")
            self.tv_durum = False
        else:
            print("Televizyon zaten kapalı...")

    def ses_ayarları(self):
        while True:
            cevap = input("Sesi açmak için '+', kısmak için '-', çıkmak için 'q' tuşuna basın:")
            if cevap == "+":
                if self.ses != 100:
                    self.ses += 1
                    print("Ses:", self.ses)
            elif cevap == "-":
                if self.ses != 0:
                    self.ses -= 1
                    print("Ses:", self.ses)
            else:
                print("Ses güncellendi:", self.ses)
                break

    def kanal_ekle(self, kanal):
        print("Kanal ekleniyor...")
        time.sleep(1)
        self.kanal_listesi.append(kanal)
        print("Kanal eklendi.")

    def kanal_sayisi(self):
        print("Kanal sayısı:", len(self.kanal_listesi))

    def rastgele_kanal(self):
        rastgele = random.randint(0, len(self.kanal_listesi)-1)
        self.kanal = self.kanal_listesi[rastgele]
        print("Şu anki kanal:", self.kanal)

    def tv_bilgileri(self):
        print("TV Durumu:", self.tv_durum)
        print("Ses:", self.ses)
        print("Kanal Listesi:", self.kanal_listesi)
        print("Şu anki kanal:", self.kanal)

k = Kumanda()

while True:
    print("""
    ==========================
    KUMANDA İŞLEMLERİ:
    
    1. TV AÇ
    2. TV KAPAT
    3. SES AYARLARI
    4. KANAL EKLE
    5. KANAL SAYISI
    6. RASTGELE KANALA GEÇ
    7. TV BİLGİLERİ
    8. ÇIKIŞ
    ==========================
    """)

    islem = input("İşlem seçiniz:")

    if islem == "1":
        k.tv_ac()
    elif islem == "2":
        k.tv_kapat()
    elif islem == "3":
        k.ses_ayarları()
    elif islem == "4":
        kanal_isimleri = input("Kanal isimlerini ',' ile ayırarak girin:")
        kanal_listesi = kanal_isimleri.split(",")
        for kanal in kanal_listesi:
            k.kanal_ekle(kanal)
    elif islem == "5":
        k.kanal_sayisi()
    elif islem == "6":
        k.rastgele_kanal()
    elif islem == "7":
        k.tv_bilgileri()
    elif islem == "8":
        print("Program sonlandırılıyor...")
        break
    else:
        print("Geçersiz işlem...")


```
